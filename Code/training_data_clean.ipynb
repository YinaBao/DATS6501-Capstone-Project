```python
def clean_text(df, col):
    n=0
    for i in df[col]:
        i = str(i)
        i = i.lower()
        i = re.sub(r'http\S+', '', i)
        i = re.sub(r'@\S+', '', i)
        i = re.sub('[^a-zA-Z]', ' ', i)
        #i = i.lower().split()
        #i = ' '.join(i)
        df[col][n]=i
        if n%10000==0:
            print('-')
        n=n+1
    return df
```

```python
print(df)
df.to_csv("selected_train.csv")
```
